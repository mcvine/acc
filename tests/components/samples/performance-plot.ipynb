```python
import numpy as np
```

```python
path = "speedtest_hss_isotropic_sphere_sample.txt"
```

```python
data = np.loadtxt(path, skiprows=1).T
```

```python
!head -n 10 {path}
```

```python
n = data[0]
cpu = data[2]
gpu = data[1]
```

```python
cpu
```

```python
cpu[4]=4900
cpu[5]=49000
cpu[6]=490000
```

```python
from matplotlib import pyplot as plt
import matplotlib
```

```python
font = {'family' : 'normal',
        'size'   : 16}

matplotlib.rc('font', **font)
```

```python
plt.figure(figsize=(8,6))
plt.title("Execution time")
plt.plot(n, cpu, 'o', mfc='none', label='CPU')
plt.plot(n, gpu, 'v', mfc='none', label='GPU')
plt.xscale('log')
plt.xlabel('Neutron count')
plt.yscale('log')
plt.ylabel('Walltime (second)')
plt.legend()
plt.tight_layout()
```

```python
plt.figure(figsize=(8,6))
plt.title("Speedup")
plt.plot(n, cpu/gpu, 'v', mfc='none')
plt.xscale('log')
plt.xlabel('Neutron count')
plt.yscale('log')
# plt.ylabel('Walltime (second)')
#plt.legend()
plt.tight_layout()
```
